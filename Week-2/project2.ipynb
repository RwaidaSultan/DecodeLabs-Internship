## 🔍 Exploratory Data Analysis (EDA)

```python
numeric_cols = ["Quantity","UnitPrice","ItemsInCart","TotalPrice"]
```

```python
product_revenue = (
    df.groupby("Product")["TotalPrice"].sum()
        .sort_values(ascending=False)
)
product_revenue
```

Chair generated the highest total revenue followed by Printer

```python
avg_order = (
    df.groupby("Product")["TotalPrice"].mean()
      .sort_values(ascending=False)
)

avg_order
```

Laptop had the highest average order value, Phone had the lowest

```python
product_orders = (
    df.groupby("Product")["OrderID"].count()
      .sort_values(ascending=False)
)
product_orders
```

Printer was the most ordered product with 181 orders

```python
payment_revenue = (
    df.groupby("PaymentMethod")["TotalPrice"].sum()
      .sort_values(ascending=False)
)
payment_revenue
```

Credit Card had the highest total revenue

```python
payment_orders = (
    df.groupby("PaymentMethod")["OrderID"].count()
      .sort_values(ascending=False)
)
payment_orders
```

Online was the most frequently used payment method with 258 orders

```python
status_revenue = (
    df.groupby("OrderStatus")["TotalPrice"].sum()
      .sort_values(ascending=False)
)
status_revenue
```

Cancelled orders had the highest status_revenue, Delivered orders had the lowest

```python
referral_orders = (
    df.groupby("ReferralSource")["OrderID"].count()
      .sort_values(ascending=False)
)
referral_orders
```

Instagram was the top referral source with 259 orders

```python
referral_revenue = (
    df.groupby("ReferralSource")["TotalPrice"].sum()
      .sort_values(ascending=False)
)
referral_revenue
```

Instagram generated the highest total revenue,Referral generated the lowest

```python
df["Month"] = df["Date"].dt.to_period("M")
monthly_revenue = (
    df.groupby("Month")["TotalPrice"].sum()
)
monthly_revenue
```

```python
monthly_orders = (
    df.groupby("Month")["OrderID"].count()
)
monthly_orders
```

### 🧐 Check Outliers

```python
numeric_cols = ["Quantity","UnitPrice","ItemsInCart","TotalPrice"]
for col in numeric_cols:
    plt.figure(figsize=(6,3))
    plt.boxplot(df[col])
    plt.title(col)
    plt.show()
```

There is an outlier in the TotalPrice column

```python
for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower) | (df[col] > upper)]

    print(f"{col}: {len(outliers)} outliers")
```
